# Embedding
Embeddings are a way to turn complex data, like words or images, into a list of numbers that a computer can understand. Think of it like translating a book into a secret code that only a computer can read.

For example, in language, each word can be represented as a point in a big space. Words that have similar meanings are placed close to each other in this space. So, "cat" and "dog" might be near each other, while "cat" and "car" would be farther apart.

This helps computers understand relationships between words, images, or other data, making it easier for them to perform tasks like translating languages, recognizing objects in pictures, or even recommending movies.

```python
import os
from openai import AzureOpenAI

openai_client: AzureOpenAI = AzureOpenAI(
    api_version = "2024-06-01",
    max_retries=5
)
```

```python
embedding = openai_client.embeddings.create(input="Hi there", model="text-embedding-ada-002")
```
